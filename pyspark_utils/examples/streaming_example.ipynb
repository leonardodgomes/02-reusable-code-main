### Streaming Module — Usage Examples
O módulo `streaming` fornece utilitários para trabalhar com Spark Structured Streaming, incluindo leitura de Kafka, Auto Loader, escrita em Delta, triggers, checkpoints e watermarks.

Essas funções são essenciais para pipelines de seguros que precisam processar:
- eventos de sinistros em tempo real
- atualizações de apólices
- logs de telemetria
- integrações com APIs externas

### Ler dados de Kafka

```python
from pyspark_utils.streaming import readers

df_kafka = readers.read_kafka(
    spark,
    servers="broker1:9092",
    topic="claims_events"
)
```

### Ler dados com Auto Loader

```python
df_auto = readers.read_autoloader(
    spark,
    path="/mnt/raw/claims",
    schema_location="/mnt/schemas/claims"
)
```

### Escrever em Delta

```python
from pyspark_utils.streaming import writers

query = writers.write_delta(
    df_auto,
    path="/mnt/curated/claims_stream",
    checkpoint="/mnt/checkpoints/claims"
)
```

### Configurar Trigger

```python
from pyspark_utils.streaming import triggers

query = triggers.trigger_processing_time(query, "30 seconds")
```

### Limpar checkpoint

```python
from pyspark_utils.streaming import checkpoints

checkpoints.clean_checkpoint("/mnt/checkpoints/claims")
```

### Adicionar Watermark

```python
from pyspark_utils.streaming import utils

df_watermarked = utils.with_watermark(df_auto, "event_time", "10 minutes")
```

### Ler Delta como Stream

```python
df_stream = readers.read_delta(spark, "/mnt/curated/claims_stream")
```
