### Common Module — Usage Examples
O módulo common fornece utilitários genéricos usados em toda a framework, incluindo:
- Decorators
- Logging
- Exceções customizadas
- Funções de data/hora
- Utilidades para DataFrames
- Carregamento de configurações

### Medir tempo de execução

```python
from pyspark_utils.common.decorators import timeit

@timeit
def process_data(df):
    return df.count()

process_data(df)
```

### Retry automático

```python
from pyspark_utils.common.decorators import retry

@retry(times=3, delay=2)
def unstable_operation():
    raise Exception("Falhou!")

unstable_operation()
```

### Logging padrão

```python
from pyspark_utils.common.logging import get_logger

logger = get_logger("pipeline")
logger.info("Iniciando pipeline...")
```

### Exceções customizadas

```python
from pyspark_utils.common.exceptions import SchemaMismatchError

raise SchemaMismatchError("Schema inválido para tabela de apólices")
```

### Utilidades de tempo

```python
from pyspark_utils.common.time_utils import timestamp_str

print("Pipeline executado em:", timestamp_str())
```

### Utilidades para DataFrames

```python
from pyspark_utils.common.dataframe_utils import show_sample

show_sample(df, n=10)
```

### Carregar configurações

```python
from pyspark_utils.common.config import load_yaml

config = load_yaml("config/pipeline.yaml")
print(config)
```
