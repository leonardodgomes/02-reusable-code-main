## IO Module — Usage Examples
The io module provides consistent, safe, and reusable helpers for reading and writing data across multiple formats.
These utilities are designed for real‑world insurance pipelines where data arrives in many shapes and must be written reliably to Delta Lake.

### Example: Reading Raw Policy Data

```python
from pyspark_utils.io import readers

df_policies = readers.read_csv(
    spark,
    path="/mnt/raw/policies.csv",
    header=True,
    infer_schema=True
)
df_policies.show()
```

### Example: Reading Claims from JSON

```python
df_claims = readers.read_json(
    spark,
    path="/mnt/raw/claims.json",
    multiline=True
)
```

### Example: Reading Delta Tables

```python
df_delta = readers.read_delta(
    spark,
    path="/mnt/curated/policies",
    version=5
)
```

## Example: Reading from JDBC

```python
df_customers = readers.read_jdbc(
    spark,
    url="jdbc:sqlserver://server:1433;database=insurance",
    table="dbo.customers",
    properties={"user": "admin", "password": "secret"}
)
```

### Writing Data
---

**Write Parquet**

```python
from pyspark_utils.io import writers

writers.write_parquet(
    df_policies,
    path="/mnt/curated/policies_parquet",
    partition_by="policy_year"
)
```

**Write Delta**

```python
writers.write_delta(
    df_claims,
    path="/mnt/curated/claims_delta",
    mode="overwrite",
    partition_by="claim_year"
)
```

**Safe Overwrite**

```python
writers.safe_overwrite(
    df_policies,
    path="/mnt/temp/policies_snapshot"
)
```

### Schema Management
---

**Load Schema from JSON**

```python
from pyspark_utils.io import schema

policy_schema = schema.load_schema_from_json("schemas/policy_schema.json")
```

**Enforce Schema**

```python
df_typed = schema.enforce_schema(df_policies, policy_schema)
```
