### Import the basics libraries

```python
import sys
import os

def set_project_root(folder_name):
    """
    Adds a project folder to sys.path so Python can import modules from it.

    Parameters
    ----------
    folder_name : str
        The name of the subfolder inside the repository that contains the project.
        Example: "python", "sql", "powershell", etc.

    How it works
    ------------
    - In a .py file:
        __file__ exists → we use it to find the script's directory.
    - In a Jupyter notebook:
        __file__ does NOT exist → we fall back to os.getcwd().

    - We then append the chosen project folder to sys.path.
    - This allows imports like:
        from calculations.stats import mean

    This function avoids hard‑coding absolute paths and works on any machine.
    """

    # Detect base directory (script or notebook)
    try:
        base_dir = os.path.dirname(os.path.abspath(__file__))
    except NameError:
        base_dir = os.getcwd()

    # Build the full path to the chosen project folder
    project_root = os.path.join(base_dir, folder_name)

    # Add to sys.path if not already present
    if project_root not in sys.path:
        sys.path.append(project_root)

```
